<font size = 4>**QUESTION 1**</font>

Design a class named Polygon that initializes with the length of a side. Then, derive a class named Square from the Polygon class. Utilize the side length defined in the Polygon class for the Square class. Within the Square class, implement a method called findArea() that calculates and returns the square's area based on its side length. Use __init__() for necessary initialization

```python
class Polygon:
    def __init__(self,s):
        self.side = s
    
    
class Square(Polygon):
    def __init__(self,s):
        super().__init__(s)
        
    def findArea(self):
        #Returns the square of the side length
        return self.side*self.side
    
S1 = Square(3)
print("Area of square with side 3 is :",S1.findArea())
```

<font size = 4>**QUESTION 2**</font>

 (a)  Create a class Father with attributes

                    - father_name (string), father_age (int), father_talents (set of strings)

                Create a class Mother with attributes:

                    - mother_name (string), mother_age (int), mother_talents (set of strings)

                 Create a class Child that inherits both father and mother with attributes: 

                    - child_name (string), child_age (int), child_gender(string)

 (b) Create a function getChildDetails() in Child to input the details of the child, it’s father and mother and printChildDetails() to print the details using     printChildDetails()

  (c) Create an object of class Child and read the details by invoking getChildDetails() and display the details entered.

  (d) Create a function displayTalents() in class Child that displays the talents of the child inherited from father and mother. A talent is inherited to a child if both

              the parents possess it.



```python
class Father:
    def __init__(self,name="none",age=-1,talents={}):
        self.father_name = name
        self.father_age = age
        self.father_talents = talents
    
class Mother:
    def __init__(self,name="none",age=-1,talents={}):
        self.mother_name = name
        self.mother_age = age
        self.mother_talents = talents
        
class Child(Father,Mother):
    def __init__(self):
        Father.__init__(self)
        Mother.__init__(self)
    
    def getChildDetails(self):
        #Below we take input for child , father and mother
        
        self.child_name = input("Child Name: ")
        self.child_age = int(input("Child Age: "))
        self.child_gender = input("Child Gender: ")
        
        self.father_name = input("Father Name: ")
        self.father_age = int(input("Father Age: "))
        self.father_talents = set(input("Father talents separated by space: ").split())
        
        self.mother_name = input("Mother Name: ")
        self.mother_age = input("Mother Age: ")
        self.mother_talents = set(input("Mother Talents separated by space: ").split())
        
        #Child talents are the intersection of father and mother
        # & is intersection operator for sets 
        self.child_talents = self.mother_talents & self.father_talents
        
        self.displayTalents()
        
    def displayTalents(self):
        #Display all child talents
        print("Child Talents:")
        for talent in self.child_talents:
            print(talent,end=" ")
        
child = Child()

print("\n")
child.getChildDetails()
        
```

<font size = 4>**QUESTION 3**</font>

  Text File Input Output

         Create a .txt (text) file and use the pledge of India as the content of the text file.

         Write a python program that reads this text file, processes it by counting the number of occurrences of each word in the file, and then writes the result back to a          new text file.5

```python
#Created text file below


pledge = "All Indians are my brothers and sisters. \
I love my country and I am proud of its \
rich and varied heritage. I shall strive to be worthy of it. \
I shall respect my parents, teachers and all elders and treat \
everyone with courtesy."
f = open("pledge.txt","w")
f.write(pledge)
f.close()
```

```python
#Reading the text file below

data = ""
with open("pledge.txt","r") as f:
    data = f.read()

wordlist = data.split()

freq = {word:0 for word in wordlist}
for word in wordlist:
    freq[word]+=1
```

```python
#Writing to text file 
#Each row has the word followed by a space and the frequency of the word

with open("wordcount.txt","w") as f:
    for key,value in freq.items():
        f.write(key+" "+str(value)+"\n")
```

<font size = 4>**QUESTION 4**</font>

For a restaurant, create a parent class ‘Bill’ which has the properties as ‘Customer  name’, ‘Table Number’ and ‘Order’ of which the name, order are strings and table  number is a positive integer. Define a method to extract the order details from the string and return a 2-D array of ordered items & their number. Create a child class ‘ 'Restaurant Bill’ which has a property called ‘Price list’ of the items and has a method to calculate the total bill amount by using the price list and order details. Also have a   method to print the complete bill for the customer including taxes and service charges.

     The strings will be of the following format:

     #Name: “Akshay” (Name of the customer)

     #Table Number: 7 (Table Number)

     #Order: “Item1x1,Item2x3,Item3x1,…” (ItemxNumber needed)

     #Price List: “Item1-100,Item2-70,Item3-250,...” (Item-Price)

```python


class Bill:
    def __init__(self,cust,table,order):
        self.customer_name = cust
        self.table_number = table
        self.order = order
    
    def extractOrder(self):
        itemlist = self.order.split(",")
        
        #For each Itemx2 we split at the last lowercase x
        #Note we aren't doing .split('x') as the item name could have x in it
        #Note that we also cannot just take the last character as the number because the number can be more that one digit long
        
        #We find the index of the last x and then slice the string accordingly
        #The string until the lastxindex - 1 is the itemname
        #The string after that x is the number 
        items_and_freq = []
        for ele in itemlist:
            last_x_index = 0
            for i in range(len(ele)):
                if(ele[i] == 'x'):
                    last_x_index = i
            
            items_and_freq += [[ele[0:last_x_index],int(ele[last_x_index+1:])]]
        
        return items_and_freq

class RestaurantBill(Bill):
    def __init__(self,cust,table,order,prices):
        super().__init__(cust,table,order)
        self.price_list = prices
    
    def Calculate(self):
        prices = self.price_list.split(",")
        price_dict = {ele.split("-")[0]: int(ele.split("-")[1]) for ele in prices}
        
        billAmt = 0
        items_and_freq = self.extractOrder()
        for ele in items_and_freq:
            billAmt += ele[1]*price_dict[ele[0]]
        
        #returns the total amount of items
        return billAmt
    
    def print_bill(self):
        # :> is syntax for keeping width and printing the variable in right side
        print(f"Bill of Restaurant")
        print(f"Customer Name:{self.customer_name:>10}")
        print(f"Table No:{self.table_number:>11}")
        print(f"ItemName UnitPrice Quantity Amount")
        
        items_and_freq = self.extractOrder() #the 2-d array of items and number
        prices = self.price_list.split(",")  #create the list items with their prices separated with -
        
        #dict of item prices created below
        price_dict = {ele.split("-")[0]: int(ele.split("-")[1]) for ele in prices}
        for ele in items_and_freq:
            print(f"{ele[0]:>8} {price_dict[ele[0]]:>9} {ele[1]:>8} {ele[1]*price_dict[ele[0]]:>6}")
        
        Taxes = self.Calculate()*0.05
        ServiceCharge = self.Calculate()*0.05
        TotalBillAmt = self.Calculate()*1.05
        
        print(f"GST(5%) {Taxes:>16}")
        print(f"Serv.Charge(5%) {ServiceCharge:>8}")
        
        print(f"Total {TotalBillAmt:>18}")
        
    def print_bill_file(self,file_name):
        bill = ""
        bill += f"Restaurant Bill\n"
        bill += f"Customer Name:{self.customer_name:>10}\n"
        bill += f"Table Number:{self.table_number:>11}\n"
        bill += f"ItemName UnitPrice Quantity Amount\n"
        
        items_and_freq = self.extractOrder()
        prices = self.price_list.split(",")
        price_dict = {ele.split("-")[0]: int(ele.split("-")[1]) for ele in prices}
        for ele in items_and_freq:
            bill += f"{ele[0]:>8} {price_dict[ele[0]]:>9} {ele[1]:>8} {ele[1]*price_dict[ele[0]]:>6}\n"
        
        Taxes = self.Calculate()*0.05
        ServiceCharge = self.Calculate()*0.05
        TotalBillAmt = self.Calculate()*1.05
        
        bill += f"GST(5%) {Taxes:>16}\n"
        bill += f"Serv.Charge(5%) {ServiceCharge:>8}\n"
        
        bill += f"Total {TotalBillAmt:>18}\n"
        
        with open(file_name,"w") as f:
            f.write(bill)
        

B1 = Bill("Rad",10,"Item1x1,Item2x3,Item3x1")
print("Extracted order example : ",B1.extractOrder())

print("\n\n")

B2 = RestaurantBill("Rad",10,"Item1x1,Item2x3,Item3x1","Item1-100,Item2-70,Item3-250")
B2.print_bill()
```
