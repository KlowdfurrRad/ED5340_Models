```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0,20,200)
x = np.cos(t)
y = np.sin(t)
z = t

#line method
fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")
ax.set_title("Helical Wave Line method")
ax.zaxis.labelpad = 1.5
ax.plot(x,y,z,color = "red",label = "Line method")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0,20,200)
x = np.cos(t)
y = np.sin(t)
z = t

#scatter method(scatter plot)
fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")
ax.set_title("Helical Wave Scatter method")
ax.zaxis.labelpad = 1.5
ax.scatter(x,y,z,label = "Scatter method")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.legend()
plt.show()
```

<font size = 4>**QUESTION 7**</font>

```python
import pandas as pd

StringData = '''Date;Event;Cost
    10/2/2011;Music;10000
    11/2/2011;Poetry;12000
    12/2/2011;Theatre;5000
    13/2/2011;Comedy;8000'''

#Split the string by \n into the different rows
rows = StringData.split("\n")
for i in range(len(rows)):
    rows[i] = rows[i].lstrip(" ") #Here we string the spaces on left side of the string as they are present here
    
labels = rows[0].split(";") #The first row has the column labels
dicti = dict()
for x in labels:
    dicti[x] = list() #For every label we create a list corresponding to that label in the dictionary

for i in range(1,len(rows)):
    elements = rows[i].split(";") #We take each row and split via ;
    for i in range(len(labels)):
        dicti[labels[i]].append(elements[i]) #For each element we append it to the appropriate list in the dictionary based on the column label
        
df = pd.DataFrame(dicti) #Create the dataframe from the dictionary
print(df)
```

```python

def first_diff(x,func,h = 0.0001):
    #f'(x) = f(x+h) - f(x-h) / 2h 
    return (func(x+h)-func(x-h))/(2*h)

def second_diff(x,func,h = 0.0001):
    #f''(x) = f(x+h) + f(x-h) - 2f(x) /h*h central difference formula for second derivative 
    return (func(x+h)+func(x-h)-2*func(x))/(h*h)

def J(w):
    return 3*w**2+2*w

print("Approximate derivative:")
for x in range(0,10):
    print(f"first derivative at {x} is {first_diff(x,J)}")
    print(f"second derivative at {x} is {second_diff(x,J)}")
```
