1. Write a program that takes coefficients A, B, C, D, and E as inputs representing a 4th degree polynomial in the form Ax^4 + Bx^3 + Cx^2 + Dx + E. Calculate the values of this polynomial for x in the range from -100 to 100, with constant discrete intervals.

Store the resulting x and y values as a NumPy array, where x represents the input values, and y represents the corresponding output values of the polynomial. Finally, use Matplotlib to plot the graph using the generated NumPy array.

```python
import numpy as np
import matplotlib.pyplot as plt

#take inputs
A,B,C,D,E = input("Give 5 coefficients separated by space").split()
A = int(A)
B = int(B)
C = int(C)
D = int(D)
E = int(E)

#given value of X finds the polynomial value
def value(A,B,C,D,E,x):
    val = 0
    val += A
    val *= x
    val += B
    val *= x
    val += C
    val *= x
    val += D
    val *= x
    val += E
    return val

#creates arrays of x and y and returns them
def poly(A,B,C,D,E):
    xlst = []
    ylst = []
    for x in range(-100,101,5):
        ylst.append(value(A,B,C,D,E,x))
        xlst.append(x)
    return np.array(xlst),np.array(ylst)

#plot the x and y arrays
x,y = poly(A,B,C,D,E)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

2. Suppose you have a dictionary containing information about monthly sales for different products over a period of time. The dictionary has the following structure.

sales_data = {

    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],

    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'Apr'],

    'Sales': [100, 150, 200, 120, 180, 220, 90, 110, 130]

}

```python
import pandas as pd

sales_data = {

    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],

    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'Apr'],

    'Sales': [100, 150, 200, 120, 180, 220, 90, 110, 130]

}

#create dataframe
df = pd.DataFrame(sales_data)

#group by the Product column
grouped = df.groupby(["Product"])

#Find sum of all sales for each product
dfgrouped = grouped.sum(numeric_only = True)
dfgrouped = dfgrouped.reset_index()

#plot the product vs sales bar graph
plt.bar(dfgrouped["Product"],dfgrouped["Sales"])
plt.xlabel("Product")
plt.ylabel("Sales per Product")
```

3. Create visualizations for the following mathematical functions using Matplotlib:

Plot the following single-variable functions over the range 

[−10,10], and include a title and labels for the axes:

(1) y = cos(x)

(2) y = e^x

(3) y = log(x), where x>0

```python
import matplotlib.pyplot as plt
import numpy as np

#Create the array of x - values
x = np.linspace(-10,10,201)

#find the cosine of each x - value
ycos = np.cos(x)
yex = np.e**x

#Create a figure with 3 axes (3 plots)
#For each we set title , xlabel and ylabel
fig,ax = plt.subplots(3,1,figsize = (5,15))
ax[0].plot(x,ycos)
ax[0].set_title("cos")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

ax[1].plot(x,yex)
ax[1].set_title("e^x")
ax[1].set_xlabel("x")
ax[1].set_ylabel("y")

x = np.linspace(0.1,10,100)
ylog = np.log(x)
ax[2].plot(x,ylog)
ax[2].set_title("log")
ax[2].set_xlabel("x")
ax[2].set_ylabel("y")
```
