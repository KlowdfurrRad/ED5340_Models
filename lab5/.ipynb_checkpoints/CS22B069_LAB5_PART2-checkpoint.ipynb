5. WAP to implement a class called "Bank_Account" representing a person's bank account.
The class should have the following attributes: account_holder_name (str), account_number(int), address (str) and balance (float).
The class should have methods to implement the following:
    deposit - Deposits a given amount into the account
    withdraw - Withdraws a given amount from the account
    check_balance - To get the current balance
    update_details - To update the name and address from the user and displays a message indicating successful update
    display_details - To display the details of the account.

```python
class Bank_Account:
    def __init__(self,name,num,addr,bal):
        self.account_holder_name = name
        self.account_number = num
        self.address = addr
        self.balance = bal
    
    def deposit(self,amount):
        self.balance += amount
    
    def withdraw(self,amount):
        self.balance -= amount
        
    def check_balance(self):
        return self.balance
    
    def update_details(self,name = "",addr = ""):
        #If the argument is default empty string then we DO NOT UPDATE
        if(name != ""):
            self.account_holder_name = name
            print("Successfully updated name:",self.account_holder_name)
        if(addr != ""):
            self.address = addr
            print("Successfully updated address to:",self.address)
    
    def display_details(self):
        print(f"Account Holder:{self.account_holder_name}")
        print(f"Account Number:{self.account_number}")
        print(f"Account address:{self.address}")
        print(f"Account Balance:{self.balance}")
        print("\n")
```

```python
account = Bank_Account("Raadhes",12345,"iitm hostel",10000)
account.display_details()

account.withdraw(1000)
account.display_details()

account.deposit(2000)
account.display_details()

account.update_details(addr = "iitm , chennai")
account.update_details(name = "Raadhes Chandaluru")
```

6.  Define a Matrix class of dimensions m X n (the values for m and n can be taken as input). Demonstrate matrix addition, subtraction, multiplication, element-by-element multiplication, scalar multiplication (use map here). Use operator overloading wherever possible. (Hint: In the constructor, use 'random' and create list of list using list comprehension. In the arguments of constructor, send the number of rows and columns)
Ensure that your implementation follows best practices for class design and encapsulation in Python. Comment your code to explain the functionality of each method.

```python
import random
import functools

class Matrix:
    def __init__(self,n,m):
        #n is number of rows
        #m is number of columns
        self.n = n
        self.m = m
        #Create matrice below by using randint function and nested list comprehension
        self.mat = [[random.randint(1,10) for j in range(m)] for i in range(n)]
        
    def __add__(self,other):
        #Check if the size of self and other matrice are same.
        #If not same print error message
        if(other.n != self.n or other.m != self.m):
            print("Matrix sizes do not match.")
            return
        
        #Add element by element below
        result = Matrix(self.n,self.m)
        for i in range(self.n):
            for j in range(self.m):
                result.mat[i][j] = self.mat[i][j] + other.mat[i][j]
        
        return result
    
    def __sub__(self,other):
        #Check if the size of self and other matrice are same.
        #If not same print error message
        if(other.n != self.n or other.m != self.m):
            print("Matrix sizes do not match.")
            return
        
        #Subtract element by element below
        result = Matrix(self.n,self.m)
        for i in range(self.n):
            for j in range(self.m):
                result.mat[i][j] = self.mat[i][j] - other.mat[i][j]
        
        return result
    
    def scalar_mul(self,k):
        result = Matrix(self.n,self.m)
        #For each row we apply map operation to multiply each element by k
        for i in range(self.n):
            result.mat[i] = list(map(lambda num : num*k , self.mat[i]))
        return result
    
    def ele_mul(self,other):
        #Check if sizes are same
        if(self.n != other.n or self.m != other.m):
            print("Matrix sizes do not match.")
            return
        
        #Multiply element wise below
        result = Matrix(self.n,self.m)
        for i in range(self.n):
            for j in range(self.m):
                result.mat[i][j] = self.mat[i][j]*other.mat[i][j]
        
        return result
    
    def __mul__(self,other):
        #For valid multiplication the number of columns of the first matrix should equal the number of rows of the second matrix
        #Check here and print error message if sizes do not comply
        if(other.n != self.m):
            print("Matrix sizes do not comply. Number of columns of first matrix must be equal to number of rows of the second matrix.")
            return
        
        #Set each element of result to zero
        result = Matrix(self.n,other.m)
        for i in range(self.n):
            for j in range(other.m):
                result.mat[i][j] = 0
        
        #Do the multiplication and store into result , use nested loops for multiplication
        for i in range(self.n):
            for j in range(other.m):
                for k in range(self.m):
                    result.mat[i][j] += self.mat[i][k]*other.mat[k][j]
        
        return result
        
    def __str__(self):
        #This generates string representation of the matrix
        #Each row is assumed converted to a string and new line is added
        return functools.reduce(lambda prev , row : str(prev)+"\n"+str(row),self.mat)+"\n"
    
    
```

```python
print("a:\n")
a = Matrix(3,4)
print(a)
print("b:\n")
b = Matrix(3,4)
print(b)

print("subtraction:\n")
print(a-b)

print("addition:\n")
print(a+b)

print("Element wise multiplication of a and b:\n")
print(a.ele_mul(b))

print("Scalar multiplication of a with 3:\n")
print(a.scalar_mul(3))


c = Matrix(3,4)
print("c matrix:\n")
print(c)
d = Matrix(4,2)
print("d matrix:\n")
print(d)

print("Multiplication of c and d:\n")
print(c*d)

```

7. Create a Python class named Time that represents a moment of time. The class should have attributes hour, minute, and second. Include the following features:
    A constructor that initializes hour, minute, and second, with validation to ensure each attribute is within its correct range (hours: 0-23, minutes: 0-59, seconds: 0-59).
    A __str__() method that returns the time in a format hh:mm:ss.
    Methods set_time(hour, minute, second) and get_time() to update and access the time, respectively.
    An add_seconds(seconds) method that adds a given number of seconds to the current time object, adjusting the hour, minute, and second attributes accordingly.

```python
class Time:
    def __init__(self,h,m,s):
        #The below logic is my decision.
        #Some may want to raise an error if out of range instead
        #Intrusctions were to check if the numbers were in bounds only
        #The specific logic was not specified hence I have made my decision on that
        
        if(h < 0):
            h = 0
        elif(h > 23):
            h = h%24
        if(m < 0):
            m = 0
        elif(m > 59):
            m = 59
        if(s < 0):
            s = 0
        elif(s > 59):
            s = 59

        self.hour = h
        self.minute = m
        self.second = s
    
    def set_time(self,h,m,s):
        if(h < 0):
            h = 0
        elif(h > 23):
            h = h%24
        if(m < 0):
            m = 0
        elif(m > 59):
            m = 59
        if(s < 0):
            s = 0
        elif(s > 59):
            s = 59

        self.hour = h
        self.minute = m
        self.second = s
        
    def get_time(self):
        #Returns the tuple of hour , minute and second
        return (self.hour,self.minute,self.second)
    
    def add_seconds(self,sec):
        self.second += sec
        if(self.second > 59):
            #If seconds overflow , add to corresponding number of minutes
            self.minute += self.second//60
            self.second = self.second%60
        
        if(self.minute > 59):
            #If minutes also overflow , add to hours
            self.hour += self.minute//60
            self.minute = self.minute%60
        
        if(self.hour > 23):
            #If hours overflow then just take modulo 24 , as it is passed to the next day
            self.hour = self.hour%24
    
    def __str__(self):
        #Note for below if the number if less than 10 , I am adding a "0" to the string
        #This is because we want hh:mm::ss representation
        time = ""
        if(self.hour < 10):
            time += "0"
        time += str(self.hour) + ":"
        
        if(self.minute < 10):
            time += "0" 
        time += str(self.minute) + ":"

        if(self.second < 10):
            time += "0" 
        time += str(self.second)
        
        return time
    

```

```python
cur = Time(4,30,15)
print("Current time:",cur)

cur.add_seconds(3750)
print("Current time after 3750 seconds:",cur)

cur.set_time(5,59,59)
print("Time after using set_time method:",cur)

print("Tuple of time with get_time method:",cur.get_time())
```

```python
import math
class Geometry:
    #BELOW i have made teh design choice to return -1 if any argument is out of bound -- (less than ZERO)
    def circle_area(r):
        if(r < 0):
            print("Invalid radius")
            return -1
        else:
            return math.pi*r*r
    def circle_perimeter(r):
        if(r < 0):
            print("Invalid radius")
            return -1
        else:
            return 2*math.pi*r
    def rectangle_area(a,b):
        if(a < 0 or b < 0):
            print("Invalid side length")
            return -1
        else:
            return a*b
    def rectangle_perimeter(a,b):
        if(a < 0 and b < 0):
            print("Invalid side length")
            return -1
        else:
            return 2*a+2*b
    def square_area(a):
        if(a < 0):
            print("Invalid side length")
            return -1
        else:
            return a*a
    def square_perimeter(a):
        if(a < 0):
            print("Invalid side length")
            return -1
        else:
            return 4*a

```

```python
print("Circle area and perimeter with radius 10:",Geometry.circle_area(10),Geometry.circle_perimeter(10))

print("Rectangle area and perimeter with sides 10,20:",Geometry.rectangle_area(10,20),Geometry.rectangle_perimeter(10,20))

print("Square area and perimeter with side 10:",Geometry.square_area(10),Geometry.square_perimeter(10))

print("Invalid arguments provided:",Geometry.square_area(-10))
```
