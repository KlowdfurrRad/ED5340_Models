2.Using steepest gradient descent, find all the local minima for the function  J(x1, x2) = (x1^2+x2−11)^2+(x1+x2^2−7)^2. While applying gradient descent, do the following (a) Fixing the value for alpha (b) use line search to determine the value for alpha. Plot the intermediate steps in the iteration to show one of the minimal point.

```python
def J(x1,x2):
    return (x1**2+x2-11)**2 + (x1+x2**2-7)**2
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plotcontour function plots the contour of J 
#Along with the gradient descent line of descent, which is passed to is as linex,liney
#The minimum point achieved plotted
#The start point of gradient descent also plotted

def plotcontour(J,linex,liney,minx1,minx2,startx1,startx2):
    #meshgrid for w1 amd w2
    x = np.linspace(-5,5,100)
    y = np.linspace(-5,5,100)
    X,Y = np.meshgrid(x,y)
    Z = np.zeros(10000).reshape(100,100)

    #create 2d array for Z also , 
    for i in range(100):
        for j in range(100):
            Z[i][j] += J(X[i][j],Y[i][j]) 
    
    #Values for which contours are drawn
    values = []
    for i in range(0,300,10):
        values += [i]
    
    #Plot the contour curve
    fig = plt.figure()
    ax = fig.add_subplot(111)
    cp = plt.contour(X,Y,Z,levels = values)
    
    #Draw the path
    plt.plot(linex,liney)
    
    #plot the minimum point with red
    plt.plot(minx1,minx2,"ro")
    plt.text(minx1,minx2,f"     Min\n({round(minx1,5)},{round(minx2,2)})")
    
    #plot the minimum point with blue
    plt.plot(startx1,startx2,"bo")
    plt.text(startx1,startx2,f"Start\n({startx1},{startx2})")

    
    plt.clabel(cp,fontsize = 10)
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title("Contour Plot")

    plt.show()
    
#plotcontour(J,[],[])
```

```python
def J_grad(x1,x2):
    dJ_dx1 = 2*(x1**2+x2-11)*(2*x1) + 2*(x1+x2**2-7)
    dJ_dx2 = 2*(x1**2+x2-11) + 2*(x1+x2**2-7)*(2*x2)
    return (dJ_dx1,dJ_dx2)

def Grad_General(x1,x2,func,h = 1e-6):
    #f'(x) = f(x+h) - f(x-h) / 2h 
    return ((func(x1+h,x2)-func(x1-h,x2))/(2*h),(func(x1,x2+h)-func(x1,x2-h))/(2*h))


#Below refines search space in a given direction
def bracket_method(s1,s2,a,b,d1,d2,J):
    #s1,s2 is initial point
    #initial a,b -- search space of alpha
    #and d1,d2 is direction to search (in case of gradient descent grad_x1 and grad_x2 as passed)
    for x in range(a+2,b,2):
        w1 = x - 2
        w2 = x
        w3 = x + 2
        #print(w1,w2,w3)
        if(J(s1 - w2*d1,s2 - w2*d2) <= J(s1 - w1*d1,s2 - w1*d2) 
           and J(s1 - w2*d1,s2 - w2*d2) <= J(s1 - w3*d1,s2 - w3*d2)):
            a = w1
            b = w3
            break
    return (a,b)

def interval_halving(s1,s2,a,b,d1,d2,J):
    #s1,s2 are the start points 
    #a,b is the interval of alpha provided
    #d1,d2 is the direction in which interval halving is done
    
    #Epsilon set to small value
    e = 1e-6

    #Interval halving done till the interval is less than epsilon
    while(b - a > e):
        L = b - a
        w1 = a + L/4
        w2 = b - L/4
        wm = (a+b)/2
    
        if(J(s1 - w1*d1,s2 - w1*d2) < J(s1 - wm*d1,s2 - wm*d2)):
            b = wm
            wm = w1
        elif(J(s1 - w2*d1,s2 - w2*d2) < J(s1 - wm*d1,s2 - wm*d2)):
            a = wm
            wm = w2
        else:
            a = w1
            b = w2
        
    return (a+b)/2

class grad_descent:
    def __init__(self,J):
        self.J = J
        self.stop_criteria = "iterations"
        self.iterations = 100
        
    def set_start(self,s1,s2):
        self.s1 = s1
        self.s2 = s2
    
    def set_stop_epsilon(self,e):
        self.stop_epsilon = e
        self.stop_criteria = "epsilon"
        
    def set_iterations(self,iterations):
        self.iterations = iterations
        self.stop_criteria = "iterations"
    
    def descend_const_alpha(self,alpha = 0.01):
        x1 = self.s1
        x2 = self.s2
        linex = [x1]
        liney = [x2]
        
        if(self.stop_criteria == "iterations"):
            for i in range(0,self.iterations):
                grad_x1,grad_x2 = J_grad(x1,x2)
                #Calculate next x1,x2
                x1 = x1 - alpha*grad_x1
                x2 = x2 - alpha*grad_x2
                #Append to the list to plot the line of descent later 
                linex.append(x1)
                liney.append(x2)
        elif(self.stop_criteria == "epsilon"):
            stop_condition = False
            while(not stop_condition):
                grad_x1,grad_x2 = J_grad(x1,x2)
                #Check if the norm of the gradient is less than stop_epsilon
                #Set thet stop_condition = True if so , then the loop will stop in this iteration
                if(np.sqrt(grad_x1**2+grad_x2**2) < self.stop_epsilon):
                    stop_condition = True
                #Calculate the next x1,x2
                x1 = x1 - alpha*grad_x1
                x2 = x2 - alpha*grad_x2
                #Append to the list to plot the line of descent later
                linex.append(x1)
                liney.append(x2)
                
        print(f"Value of J at the minimal point {x1,x2} is:",J(x1,x2))
        plotcontour(J,linex,liney,x1,x2,self.s1,self.s2)
        return (x1,x2)
        
    
    def descend_vary_alpha(self):
        x1 = self.s1
        x2 = self.s2
        linex = [x1]
        liney = [x2]
        
        if(self.stop_criteria == "iterations"):
            for i in range(0,self.iterations):
                
                #!!!!! Here I am using J_grad which will only work for the function provided
                #If we need to use general grad for any function then use the one I made above
                grad_x1,grad_x2 = J_grad(x1,x2)
                
                #Initially thought of normalising grad_x1,grad_x2 but then it was giving better results without that
                
                #find the refined alpha region
                alpha1,alpha2 = bracket_method(x1,x2,0,200,grad_x1,grad_x2,J)
                
                #Interval halving to find the best alpha in that direction
                alpha_best = interval_halving(x1,x2,alpha1,alpha2,grad_x1,grad_x2,J)
                
                
                #Calculate the next x1,x2
                x1 = x1 - alpha_best*grad_x1
                x2 = x2 - alpha_best*grad_x2
                
                #append the x1,x2 to list to plot the line of descent later
                linex.append(x1)
                liney.append(x2)
                
        elif(self.stop_criteria == "epsilon"):
            
            stop_condition = False
            while(not stop_condition):
                #Calculate norm of the grad, sir mentionned stop condition 
                # is to be checked with norm of grad at start of loop
                grad_x1,grad_x2 = J_grad(x1,x2)
                if(np.sqrt(grad_x1**2+grad_x2**2) < self.stop_epsilon):
                    stop_condition = True
                
                #find refined search space of alpha1,alpha2
                alpha1,alpha2 = bracket_method(x1,x2,0,200,grad_x1,grad_x2,J)
                
                #Interval halving to find the best alpha in that direction
                alpha_best = interval_halving(x1,x2,alpha1,alpha2,grad_x1,grad_x2,J)
                
                #Calculate the next x1,x2
                x1 = x1 - alpha_best*grad_x1
                x2 = x2 - alpha_best*grad_x2
                
                #append the x1,x2 to list to plot the line of descent later
                linex.append(x1)
                liney.append(x2)
                
        print(f"Value of J at the minimal point {x1,x2} is:",J(x1,x2))
        plotcontour(J,linex,liney,x1,x2,self.s1,self.s2)
        
        return (x1,x2)    

```

I have created a class named **grad_descent**.\

It two **STOP Criteria Modes** : (Default is 100 iterations) \
    - Number of **Iterations** (set_iterations method is used to set the number of iterations) \
    - **Norm of gradJ** falls below a certain epsilon (set_stop_epsilon method is used to set it) \

There are two functions to do gradient descent. \
    - grad_const_alpha : Gradient descent with a constant alpha \
        NOTE that above you can pass the constant alpha that you prefer when calling it. Default is 0.01. \
    - grad_vary_alpha : Gradient descent with alpha determined by line search method. (Bracketing and then interval halving ) \

<font size = 4>Part A</font>

1. Gradient descent with constant alpha of 0.01 \
And set iterations as 100

NOTE : The default ALPHA for constant alpha descent is set as 0.01 in the class function descent_const_alpha

```python
model = grad_descent(J)
model.set_iterations(100)
model.set_start(0,0)
model.descend_const_alpha()
```

2. Gradient descent with constant alpha of 0.01 \
And stop gradient descent when the norm of gradJ falls below 10^(-10)

```python
model = grad_descent(J)
model.set_stop_epsilon(1e-10)
model.set_start(0,0)
model.descend_const_alpha()
```

3. Gradient descent with constant alpha = 0.001 and 100 iterations \
But start point changed to (-5,0)

```python
model = grad_descent(J)
model.set_iterations(1000)
model.set_start(-5,0)
model.descend_const_alpha(1e-3)
```
